```python
import pandas as pd

combined = pd.read_csv("combined.csv")

# Duration in minutes
combined['duration_min'] = combined['Total duration (ms)'] / 60000.0  # ms â†’ minutes

# Remove impossible / extreme journeys
combined = combined[(combined['duration_min'] >= 1) & (combined['duration_min'] <= 120)]

# 1. Stats per bike type
stats_by_type = combined.groupby("Bike model")["duration_min"].describe()

# 2. Overall stats (all bikes together)
overall = combined["duration_min"].describe().to_frame().T
overall.index = ["Overall"]

# 3. Combine type stats + overall
full_stats = pd.concat([stats_by_type, overall])

# 4. Compute mode per bike type
mode_by_type = combined.groupby("Bike model")["duration_min"].agg(
    lambda x: x.mode().iloc[0] if not x.mode().empty else pd.NA
)

# 5. Compute overall mode
overall_mode_series = combined["duration_min"].mode()
overall_mode = overall_mode_series.iloc[0] if not overall_mode_series.empty else pd.NA

# 6. Add mode as a new column
full_stats["mode"] = pd.concat(
    [mode_by_type, pd.Series({"Overall": overall_mode})]
)

# 7. Transpose the table (stats as rows, bike types + overall as columns)
full_stats_T = full_stats.T.round(2)


print("\n=== Full stat data (rows = statistics, columns = bike types + overall) ===")
print(full_stats_T)
```
